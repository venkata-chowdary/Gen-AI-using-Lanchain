```python
import random
class NakliLLM:
    def __init__(self):
        print("NakliLLM initialized")
        
    def predict(self, prompt):
        print(f"NakliLLM received prompt: {prompt}")
        response_list=["NakliLLM response 1", "NakliLLM response 2", "NakliLLM response 3"]
        
        return {'response': random.choice(response_list)}
```

```python
llm=NakliLLM()
```

```python
llm.predict("Hello, how are you?")
```

```python
class NakliPromptTemplate:
    def __init__(self, template, input_variables):
        self.template = template
        self.input_variables = input_variables
        
    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
template=NakliPromptTemplate(
    template="What is the capital of {country}?",
    input_variables=["country"]
)
```

```python
template.format({"country": "France"})
```

```python
#these chain classes are not flexible, 
class NakliLLMChain:
    def __init__(self, llm, prompt):
        self.llm=llm
        self.promp=prompt
    
    def run(self, input_dict):
        formatted_prompt=self.promp.format(input_dict)
        print(f"Formatted Prompt: {formatted_prompt}")
        return self.llm.predict(formatted_prompt)
```

```python
chain=NakliLLMChain(llm=llm, prompt=template)
chain.run({"country": "Germany"})
```
