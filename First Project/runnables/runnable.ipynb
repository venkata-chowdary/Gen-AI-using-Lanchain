```python
from abc import ABC, abstractmethod
```

```python
class Runnable(ABC):
    
    @abstractmethod
    def invoke(input_data):
        pass
```

```python
import random
class NakliLLM(Runnable):
    def __init__(self):
        print("NakliLLM initialized")
        
    def  invoke(self,prompt):
        print(f"NakliLLM received prompt: {prompt}")
        response_list=["NakliLLM response 1", "NakliLLM response 2", "NakliLLM response 3"]
        
        return {'response': random.choice(response_list)}
    
    def predict(self, prompt):
        print(f"NakliLLM received prompt: {prompt}")
        response_list=["NakliLLM response 1", "NakliLLM response 2", "NakliLLM response 3"]
        
        return {'response': random.choice(response_list)}
    #here predict is deprecated, invoke will do same thing of predict
```

```python
#cant intialize object without defining abstract funtion which are inherited from the abstract class
llm=NakliLLM()
```

```python
class NakliPromptTemplate(Runnable):
    def __init__(self, template, input_variables):
        self.template = template
        self.input_variables = input_variables
    def invoke(self, input_dict):
        return self.template.format(**input_dict)
    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
class RunnableConnector(Runnable):
    def __init__(self, runnable_list):
        self.runnable_list=runnable_list
        
    def invoke(self, input_data):
        for runnable in self.runnable_list:
            #this will execute prompt.invoke(), llm.invoke()
            input_data=runnable.invoke(input_data)

        return input_data
```

```python
class NakliPromptTemplate:
    def __init__(self, template, input_variables):
        self.template = template
        self.input_variables = input_variables
        
    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
template=NakliPromptTemplate(
    template="What is the capital of {country}?",
    input_variables=["country"]
)
```

```python
class NakliStrOpParser(Runnable):
    def __init__(self):
        pass
    
    def invoke(self, input_data):
        return input_data['response']
```

```python
parser=NakliStrOpParser()
```

```python
chain=RunnableConnector([template,llm])
print(chain.invoke({"topic": "Cricket"}))
```
